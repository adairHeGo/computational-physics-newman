Exercise 2.1: Another ball dropped from a tower

A ball is again dropped from a tower of height $h$ with initial velocity 0. Write a program  that asks  the user  to enter the height in meters of the tower and then calculates and prints  the time the ball takes until it hits the ground, ignoring air resistance. Use your program to calculate the time for a ball dropped from a 100 m high tower. 

Using Newton's second law:
$$ma=-mg,$$
we solve the differential equation integrating over $t$ and using the initial conditions $v_0=0$ and $z_0=h$. So
$$z=-\frac{1}{2}g t^2 + h.$$
We are asked to calculate the time it takes the ball to hit the ground, this means $z(\tau)=0$:
$$-h=-\frac{1}{2}g\tau^2,$$
solving for $\tau$:
$$\tau=\sqrt{\frac{2h}{g}}$$


```python
#We need h
h = float(input("Enter h: "))
#acceleration of gravity
g = 9.81
#We calculate t
t=(2*h/g)**(1/2)
print(t)
```

A satellite is to be launched into a circular orbit around the Earth so that it orbits the planet once every $T$ seconds.
a) Show that the altitude $h$ above the Earth's surface that the satellite must have is 
$$h=\left(\frac{GMT^2}{4\pi^2}\right)^{1/3}-R,$$
where $G=6.67\times 10^{-11}m^3kg^{-1}s^{-2}$ is Newton's gravotational constant, $M=5.97\times 10^{24}$ kg is the mass of the Earth, and $R=6371$ km is its radius.

We use Newton's second law:
$$ma=G\frac{Mm}{r^2}.$$
For a circular orbit with period $T$:
$$a=r\omega^2=r \left(\frac{2\pi}{T}\right)=\frac{r 4pi^2}{T^2}.$$
So,
$$r=\left(\frac{GMT^2}{4\pi^2}\right)^{1/3},$$
or
$$h=\left(\frac{GMT^2}{4\pi^2}\right)^{1/3}-R.$$


b) Write a program that asks the user to enter the desired value of $T$ and then calculates and prints out the correct altitude in meters 

```python
#We define the constants
G = 6.67*(10**(-11))
M = 5.97*(10**(24))
R = 6371000 #in meters
pi = 3.1416
#We ask for T
T = float(input("Enter the desired value of T: "))
#We calculate h
h = ((G*M*(T**2))/(4*(pi**2)))**(1/3) - R
#Print h
print(h, "m")
```

Exercise 2.5: Quantum potential step
A well known quantum mechanics problem involves a particle of mass $m$ that encounters a one-dimensional potential step, like this: (figure in the book)
The particle with initial kinetic energy $E$ and wavevector $k_1=\sqrt(2mE)/\hbar$ enters from the left and encounters a sudden jump in potential energy of height $V$ at position $x=0$. By solving the Schr\"{o}dinger equation, one can show thet when $E>V$ the particle may either (a) pass the step, in which case it has a lower kinetic energy $E-V$ on the other side and a correspondingly smaller wavevector of $k_2=\sqrt{2m(E-V)}/\hbar$, or (b) it may be reflected , keeping all  of its kinetic energy and an unchanged wavevector but moving in the opposite direction. The probabilities $T$ and $R$ for transmission and reflection are given by
$$T=\frac{4k_1k_2}{(k_1+k_2)^2}, R=\left(\frac{k_1-k_2}{k_1+k_2}\right)^2.$$
Suppose we have a particle with mass equal to the electron mass $m=9.11\times 10^{-31}$ kg and energy 10 eV encountering a potential step of height 9 eV. Write a python program to compute and print out the transmission and reflection probabilities using the formulas above.

```python
m = 9.11*(10**(-31))
E = 10 
V = 9
hbar = 6.582*(10**(-16))
# k1 and k2
from math import sqrt
k1 = sqrt(2*m*E)/hbar
k2 = sqrt(2*m*(E-V))/hbar
#T and R
T = (4*k1*k2)/((k1+k2)**2)
R = ((k1-k2)/(k1+k2))**2
print("T=",T)
print("R=",R)
```
