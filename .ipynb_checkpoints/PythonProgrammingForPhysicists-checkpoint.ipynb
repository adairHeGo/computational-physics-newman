Exercise 2.1: Another ball dropped from a tower

A ball is again dropped from a tower of height $h$ with initial velocity 0. Write a program  that asks  the user  to enter the height in meters of the tower and then calculates and prints  the time the ball takes until it hits the ground, ignoring air resistance. Use your program to calculate the time for a ball dropped from a 100 m high tower. 

Using Newton's second law:
$$ma=-mg,$$
we solve the differential equation integrating over $t$ and using the initial conditions $v_0=0$ and $z_0=h$. So
$$z=-\frac{1}{2}g t^2 + h.$$
We are asked to calculate the time it takes the ball to hit the ground, this means $z(\tau)=0$:
$$-h=-\frac{1}{2}g\tau^2,$$
solving for $\tau$:
$$\tau=\sqrt{\frac{2h}{g}}$$


```python
#We need h
h = float(input("Enter h: "))
#acceleration of gravity
g = 9.81
#We calculate t
t=(2*h/g)**(1/2)
print(t)
```

Exercise 2.2
A satellite is to be launched into a circular orbit around the Earth so that it orbits the planet once every $T$ seconds.
a) Show that the altitude $h$ above the Earth's surface that the satellite must have is 
$$h=\left(\frac{GMT^2}{4\pi^2}\right)^{1/3}-R,$$
where $G=6.67\times 10^{-11}m^3kg^{-1}s^{-2}$ is Newton's gravotational constant, $M=5.97\times 10^{24}$ kg is the mass of the Earth, and $R=6371$ km is its radius.

We use Newton's second law:
$$ma=G\frac{Mm}{r^2}.$$
For a circular orbit with period $T$:
$$a=r\omega^2=r \left(\frac{2\pi}{T}\right)=\frac{r 4pi^2}{T^2}.$$
So,
$$r=\left(\frac{GMT^2}{4\pi^2}\right)^{1/3},$$
or
$$h=\left(\frac{GMT^2}{4\pi^2}\right)^{1/3}-R.$$


b) Write a program that asks the user to enter the desired value of $T$ and then calculates and prints out the correct altitude in meters 

```python
#We define the constants
G = 6.67*(10**(-11))
M = 5.97*(10**(24))
R = 6371000 #in meters
pi = 3.1416
#We ask for T
T = float(input("Enter the desired value of T: "))
#We calculate h
h = ((G*M*(T**2))/(4*(pi**2)))**(1/3) - R
#Print h
print(h, "m")
```

Exercise 2.3: Write a program to perform the inverse operation to that of Example 2.2. That is, ask the user for the cartesian coordinates $x,y$ of a point in two dimensional space, and calculate and print the corresponding polar coordinates, with the angle $\theta$ given in degrees.

We have:
$$r^2=x^2+y^2,$$
and 
$$\frac{y}{x}=\tan\theta.$$
So we have the special case when $x=0 \implies \theta=\pi/2, 3\pi/2$. Of course we have to add that $\tan \theta$ is not one to one in the interval $\theta \in [0,2\pi).$ So we have:
$$\theta=\arctan \frac{y}{x}, x> 0, y>0$$
$$\theta=\arctan \frac{y}{x}+\pi, x< 0, y>0$$
$$\theta=\arctan \frac{y}{x}+\pi, x> 0, y>0$$
$$\theta=\arctan \frac{y}{x}+2\pi, x> 0, y>0$$
$$\theta=\pi/2, x=0,y>0$$
$$\theta=3\pi/2, x=0, y<0$$



```python
from math import sqrt, atan, pi
#We ask for the x and y coordinates
x = float(input("Introduce the x coordinate: "))
y = float(input("Introduce the y coordinate: "))
#We calculate r
r = sqrt(x**2 + y**2)
#Now we calculate theta. So in the text we haven't seen if statements and that kind of stuff so will use atan
theta = atan(y/x) #with x =! 0
print(r)
print(theta*180/pi)
```
