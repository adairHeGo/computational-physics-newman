Exercise 7.1: Fourier transforms of simple functions

```python
import numpy as np
from cmath import exp, pi

def dft(y):
    N = len(y)
    c = np.zeros(N//(2+1), complex)
    for k in range(N//(2+1)):
        for n in range(N):
            c[k] += y[n]*exp(-2j*pi*k*n/N)
    return c
```

```python
import matplotlib.pyplot as plt
#a)
N = 1000
x = np.linspace(0,10, N)
y_a = np.zeros(N)
for i in range(N//2):
    y_a[i] = 1
    
plt.plot(x, y_a)
```

```python
c = dft(y_a)
plt.plot(abs(c))
```

```python
#b)
y_b = [n for n in range(N)]
plt.plot(x, y_b)
```

```python
c = dft(y_b)
plt.plot(abs(c))
```

```python
from cmath import sin
#c)
y_c = [sin(pi*n/N)*sin(20*pi*n/N) for n in range(N)]
plt.plot(x, y_c)
```

```python
c = dft(y_c)
plt.plot(abs(c))
```

```python
def f(t):
    if ((2*t)//2)%2 == 0:
        return 1
    else:
        return -1
```

```python
import numpy as np

t = np.linspace(0,2,1000)
f_t = [f(ti) for ti in t]
```

```python
c = np.fft.rfft(f_t)
for i in range(10, len(c)):
    c[i] = 0
    
c_i = np.fft.irfft(c)
```

```python
import matplotlib.pyplot as plt

plt.plot(t, f_t)
plt.plot(t, c_i)
```

```python
import numpy as np
import matplotlib.pyplot as plt

dow = np.loadtxt('dow2.txt')

plt.plot(dow)
```

```python
c = np.fft.rfft(dow)

for i in range(len(c)//50, len(c)):
    c[i] = 0
    
c_i = np.fft.irfft(c)
```

```python
import matplotlib.pyplot as plt

plt.plot(dow)
plt.plot(c_i)
```

```python
def dct(y):
    N = len(y)
    y2 = np.empty(2*N, float)
    for n in range(N):
        y2[n] = y[n]
        y2[2*N-1-n] = y[n]
    c = np.fft.rfft(y2)
    phi = np.exp(-1j*np.pi*np.arange(N)/(2*N))
    return np.real(phi*c[:N])
```

```python
cc = dct(dow)
for i in range(len(cc)//50, len(cc)):
    cc[i] = 0
cc_i = np.fft.irfft(cc)
```

```python
plt.plot(dow)
plt.plot(cc_i)
plt.xlim(0, len(dow))
```

Exercise 7.7: Fast Fourier Transform

```python
import numpy as np
from cmath import exp
data = np.loadtxt('pitch.txt')

N = len(data)
total_stages = int(np.log2(N))
c = np.zeros(N//(2+1), complex)
for k in range(N//(2+1)):
    Emj_k = np.zeros(N, complex)
    Emplusonej_k = np.zeros(N, complex)
    for j in range(2**total_stages):
        Emplusonej_k[j] = data[j]#*exp(1j*2*np.pi*k*j/N) 
    
    for m in range(total_stages-1, -1, -1):
        for j in range(2**m):
            Emj_k[j] = Emplusonej_k[j] + exp(-1j*2*np.pi*(2**m)*k/N)*Emplusonej_k[j+(2**m)]
        Emplusonej_k = Emj_k.copy()

    c[k] = Emj_k[0]        

```

```python
import matplotlib.pyplot as plt
plt.plot(data)
```

```python
plt.plot(abs(c))
```

```python
c_n = np.fft.rfft(data)
plt.plot(abs(c_n))
```

```python
c_n = np.fft.rfft(data)
plt.plot(abs(c_n))
plt.plot(abs(c))
```

Exercise 7.8: Diffraction gratings

```python
from math import sin, pi
import numpy as np
import matplotlib.pyplot as plt

alpha = pi/(20e-6)
w = 200e-6
W = 10*w
wavel = 500e-9
f = 1

def q(u):
    return sin(alpha*u)**2

N = 400
u_n = [(n*W/N)-W/2 for n in range(N)]
y_n = np.array([q(ui) for ui in u_n])

ck = np.fft.rfft(y_n)
x_k = [(wavel*f/W)*k for k in range(len(ck))]
I_x = [((W**2)/N**2)*(abs(ck_i)**2) for ck_i in ck]

plt.plot(x_k, I_x)
```

Exercise 7.9: Image deconvolution

```python
import numpy as np
import matplotlib.pyplot as plt

b = np.loadtxt("blur.txt")

plt.imshow(b)
```

```python
from math import exp, cos, sin, pi
N = len(b)
sigma = 25

def f(x, y):
    return exp(-(cos(x)**2 + cos(y)**2)/(2*(sigma**2)))

f_xy = np.zeros((N, N))
x = np.linspace(-pi/2, pi/2,N)

for i in range(N):
    for j in range(N):
        f_xy[i,j] = f(x[i], x[j])

plt.imshow(f_xy, 'gray')
```

```python
from math import exp, cos, sin, pi
N = len(b)
sigma = 25

def f(x, y):
    return exp(-(x**2 + y**2)/(2*(sigma**2)))

f_xy = np.zeros((N, N))
x = np.linspace(-1, 1, N)

for i in range(N):
    for j in range(N):
        f_xy[i,j] = f(-N/2 + (i+N/2)%N, -N/2 + (j+N/2)%N)

plt.imshow(f_xy, 'gray')
```

```python
b_kl = np.fft.rfft2(b)
f_kl = np.fft.rfft2(f_xy)
a_kl = np.zeros_like(f_kl)

for i in range(len(f_kl)):
    for j in range(len(f_kl[0])):
        if abs(f_kl[i,j]) < 1e-4:
            a_kl[i,j] = b_kl[i,j]
        else:
            a_kl[i,j] = (b_kl[i,j]/f_kl[i,j])*(1/16)

a = np.fft.irfft2(a_kl)

plt.imshow(a)
```
