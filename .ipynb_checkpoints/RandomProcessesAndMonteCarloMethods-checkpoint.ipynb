Exercise 10.2: Radioactive decay chain

```python
from random import random
import numpy as np
import matplotlib.pyplot as plt

#Constants
NPb_209 = 0
NBi_209 = 0
NTl_209 = 0
NBi_213 = 10000
tau1 = 3.3*60
tau2 = 2.2*60
tau3 = 46.0*60
h = 1.0
p1 = 1 - 2**(-h/tau1)
p2 = 1 - 2**(-h/tau2)
p3 = 1 - 2**(-h/tau3)
tmax = 20000

#List of plot points
tpoints = np.arange(0.0,tmax,h)
Pb_209points = []
Bi_209points = []
Tl_209points = []
Bi_213points = []

#Main loop
for t in tpoints:
    Pb_209points.append(NPb_209)
    Bi_209points.append(NBi_209)
    Tl_209points.append(NTl_209)
    Bi_213points.append(NBi_213)
    
    decay = 0
    for i in range(NPb_209):
        if random()<p1:
            decay += 1
    NPb_209 -= decay
    NBi_209 += decay
    
    decay = 0
    for i in range(NTl_209):
        if random()<p2:
            decay += 1
    NTl_209 -= decay
    NPb_209 += decay
    
    decay1 = 0
    decay2 = 0
    for i in range(NBi_213):
        if random()<p3:
            if random()<0.9791:
                decay1 += 1
            else:
                decay2 += 1
    NBi_213 -= decay1
    NPb_209 += decay1
    NBi_213 -= decay2
    NTl_209 += decay2
    
plt.plot(tpoints, Pb_209points)
plt.plot(tpoints, Bi_209points)
plt.plot(tpoints, Tl_209points)
plt.plot(tpoints, Bi_213points)
```

Exercise 10.4: Radioactive decay again

```python
import numpy as np
import random as rd
import matplotlib.pyplot as plt

NTl = []
for n in range(1000):
    r = rd.random()
    NTl.append(-((60*3.053)/np.log(2))*np.log(1-r))
    
plt.plot(np.sort(NTl))
```

Exercise 10.5

```python
from math import sin, sqrt
import random as rd

def f(x):
    return sin(1/(x*(2-x)))**2

N = 10000
points = 0

for i in range(N):
    x = 2*rd.random()
    y = rd.random()
    if y<f(x):
        points += 1
        
I = 2*points/N
        
print(I)
print(sqrt(I*(2-I))/sqrt(N))
```

```python
N = 10000
h = 2/N
I = 0
f_2 = 0

for i in range(N):
    x = 2*rd.random()
    I += f(x)
    f_2 += f(x)**2    
print(I*h)
I /= N
f_2 /= N
print(h*sqrt(f_2 - I**2)/sqrt(N))
```
