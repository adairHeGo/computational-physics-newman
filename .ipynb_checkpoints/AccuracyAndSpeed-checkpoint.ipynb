Exercise 4.2: Quadratic equation

```python
from math import sqrt
# a)
def quad_sol(a, b, c):
    root = sqrt(b**2 - 4*a*c)
    return (-b-root)/(2*a), (-b+root)/(2*a)

quad_sol(0.001, 1000, 0.001)
```

b)

$$\frac{-b\pm\sqrt{b^2-4ac}}{2a}\times \frac{-b\mp \sqrt{b^2-4ac}}{-b\mp \sqrt{b^2-4ac}}=\frac{b^2 - b^2+4ac}{2a(-b\mp\sqrt{b^2-4ac})}=\frac{2c}{-b\mp \sqrt{b^2-4ac}}$$

```python
def quad_sol(a, b, c):
    root = sqrt(b**2 - 4*a*c)
    return (2*c)/(-b+root), (2*c)/(-b-root)

quad_sol(0.001, 1000, 0.001)
```

I have more digits for the positive root with the second method and less digits for the negative root compared to the first method.

```python
# c)
def quad_sol(a, b, c):
    root = sqrt(b**2 - 4*a*c)
    return (2*c)/(-b+root), (-b+root)/(2*a)
```

```python
quad_sol(0.001, 1000, 0.001)
```

Exercise 4.3: Calculating derivatives

```python
def f(x):
    return x*(x-1)

delta = 0.01

(f(1+delta)-f(1))/delta
```

$$f'(x)=x + x-1 =2x-1$$
$$f'(1)=2-1=1$$

```python
import numpy as np
import matplotlib.pyplot as plt
delta = np.array([10e-2, 10e-4, 10e-6, 10e-8,10e-10,10e-12,10e-14,10e-16,10e-18])
derivative = (f(1+delta)-f(1))/delta

plt.plot(delta, derivative)
```

The reason for this is because we're dividing over $\delta$.

Exercise 4.4: Calculating integrals

```python
from math import sqrt, pi
N = 100000
h = 2/N
I = 0

for k in range(N):
    x_k = -1 + h*k
    y_k = sqrt(1 - x_k**2)
    
    I += h*y_k
    
print(I)
print(pi/2)
```
