```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n*factorial(n-1)
```

Exercise 4.2: Quadratic equation

```python
from math import sqrt
# a)
def quad_sol(a, b, c):
    root = sqrt(b**2 - 4*a*c)
    return (-b-root)/(2*a), (-b+root)/(2*a)

quad_sol(0.001, 1000, 0.001)
```

$$\frac{-b\pm\sqrt{b^2-4ac}}{2a}\times \frac{-b\mp \sqrt{b^2-4ac}}{-b\mp \sqrt{b^2-4ac}}=\frac{b^2 - b^2+4ac}{2a(-b\mp\sqrt{b^2-4ac})}=\frac{2c}{-b\mp \sqrt{b^2-4ac}}$$

```python
def quad_sol(a, b, c):
    root = sqrt(b**2 - 4*a*c)
    return (2*c)/(-b+root), (2*c)/(-b-root)

quad_sol(0.001, 1000, 0.001)
```
