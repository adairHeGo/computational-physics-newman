Exercise 8.1: A low-pass filter

```python
import numpy as np
import matplotlib.pyplot as plt

RC = 0.01

def f(Vout,t):
    if (int(2*t)%2) == 0:
        return (1 - Vout)/RC
    else:
        return (-1 - Vout)/RC
    
t = np.linspace(0, 10, 1000)
V0 = 0
VOpoints = []
h = t[1]-t[0]

for ti in t:
    VOpoints.append(V0)
    k1 = h*f(V0, ti)
    k2 = h*f(V0 + 0.5*k1, ti + 0.5*h)
    k3 = h*f(V0 + 0.5*k2, ti + 0.5*h)
    k4 = h*f(V0 + k3, ti + h)
    V0 += (k1 + 2*k2 + 2*k3 + k4)/6

plt.plot(t, VOpoints, '.')
        
```

Exercise 8.2: The Lotka-Volterra equations

```python
import numpy as np
import matplotlib.pyplot as plt

alpha = 1
beta = 0.5
gamma = 0.5
delta = 2

def f(r, t):
    x = r[0]
    y = r[1]
    fx = alpha*x - beta*x*y
    fy = gamma*x*y - delta*y
    return np.array([fx,fy], float)

r = np.array([2.0, 2.0], float)

tpoints = np.linspace(0, 30, 100)
xpoints = []
ypoints = []
h = tpoints[1] - tpoints[0]

for t in tpoints:
    xpoints.append(r[0])
    ypoints.append(r[1])
    k1 = h*f(r, t)
    k2 = h*f(r + 0.5*k1, t + 0.5*h)
    k3 = h*f(r + 0.5*k2, t + 0.5*h)
    k4 = h*f(r + k3, t + h)
    r += (k1 + 2*k2 + 2*k3 + k4)/6

plt.plot(tpoints, xpoints, '-')
plt.plot(tpoints, ypoints, '-')
```

```python
plt.plot(xpoints, ypoints, '.')
plt.xlabel('rabbits')
plt.ylabel('foxes')
```

Exercise 8.3: The Lorenz equations

```python
import numpy as np
import matplotlib.pyplot as plt

sigma = 10
r = 28
b = 8/3

def f(v, t):
    x = v[0]
    y = v[1]
    z = v[2]
    fx = sigma*(y - x)
    fy = r*x - y - x*z
    fz = x*y - b*z
    return np.array([fx,fy,fz], float)

v = np.array([0.0, 1.0, 0.0], float)

tpoints = np.linspace(0, 50, 1000)
xpoints = []
ypoints = []
zpoints = []
h = tpoints[1] - tpoints[0]

for t in tpoints:
    xpoints.append(v[0])
    ypoints.append(v[1])
    zpoints.append(v[2])
    k1 = h*f(v, t)
    k2 = h*f(v + 0.5*k1, t + 0.5*h)
    k3 = h*f(v + 0.5*k2, t + 0.5*h)
    k4 = h*f(v + k3, t + h)
    v += (k1 + 2*k2 + 2*k3 + k4)/6

plt.plot(tpoints, ypoints, '.')
```
